```python
# Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
```

```python
# Read the free energy values and convert the time
# and the free energies to numpy arrays
df = pd.read_csv("FreeEnergy_phi_zero_N_50.csv")
time = df["Time "].to_numpy()
F = df["Free Energy "].to_numpy()
```

```python
plt.figure(figsize=(8,8))
plt.plot(time, F)
sns.set_theme()
plt.title("Free Energy Density of the system against time")
plt.xlabel("Time")
plt.ylabel("Free Energy")
plt.show()
```
